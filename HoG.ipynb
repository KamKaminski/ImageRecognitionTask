```python
from skimage import feature
from sklearn import datasets
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import numpy as np
import matplotlib.pyplot as plt
lfw_people = datasets.fetch_lfw_people(min_faces_per_person=20, resize=0.4)
def compute_HOG(image):
    # Perform HOG feature extraction using skimage library
    H = feature.hog(image, orientations=9, pixels_per_cell=(8, 8),
                     cells_per_block=(2, 2), transform_sqrt=True, block_norm="L2-Hys")
    
    return H

# the HOG features and labels
data = []
labels = []

for image, label in zip(lfw_people.images, lfw_people.target):
    H = compute_HOG(image)
    data.append(H)
    labels.append(label)

# Turnig our variables to numpy arrays
data = np.array(data)
labels = np.array(labels)

# Split the data into training and test sets
(trainData, testData, trainLabels, testLabels) = train_test_split(data, labels, test_size=0.25, random_state=42, stratify=labels)

# SVM classifier
model = svm.SVC(kernel='poly', probability=True, random_state=42)

# Train the classifier with training data
model.fit(trainData, trainLabels)

# Make predictions
predictions = model.predict(testData)

# Find unique labels
unique_test_labels = np.unique(testLabels)

# only includes labels present in test data
new_target_names = [lfw_people.target_names[i] for i in unique_test_labels]
print(classification_report(testLabels, predictions, target_names=lfw_people.target_names, zero_division=0))
```
